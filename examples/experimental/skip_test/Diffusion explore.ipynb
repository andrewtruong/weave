```python
import weave
from weave import weave_internal
from weave.ecosystem.all import langchain
weave.use_frontend_devmode()
```

```python
weave.panels.Board(
    vars={
        "chat_model": langchain.chat_openai("gpt-3.5-turbo", 0.7),
        "topic": "animals",
        "num_cols": 2,
        "col_names": lambda chat_model, num_cols: weave.ops.range(0, num_cols, 1).map(
            lambda i: chat_model.predict(weave_internal.const("You are a creative writer. You've come up with ")
                                         + i.toString()
                                         + " ordered descriptive phrases (3 words or fewer) that represent how magical something is, from least magical to most. Phrase "
                                         + i.toString()
                                         + " is:")
        ),
        "num_rows": 2,
        "row_names": lambda chat_model, topic, num_rows: weave.ops.range(0, num_rows, 1).map(
            lambda i: chat_model.predict(weave_internal.const("You are a creative writer. You've come up with ")
                                         + i.toString()
                                         + " " + topic
                                         + " that are unlikely to be seen together. Number "
                                         + i.toString()
                                         + " is: ")
        ),

    },
    panels=[
        weave.panels.BoardPanel(
            lambda num_rows, num_cols, row_names, col_names: weave.panels.Facet(
                weave.ops.range(0, num_rows * num_cols, 1),
                cell_size=(192, 192),
                x=lambda i: col_names[i % num_cols],
                y=lambda i: row_names[(i / num_cols).floor()],
                select=lambda cell_values: (
                    weave_internal.const("cute hd cartoon of ")
                    + row_names[(cell_values[0] / num_cols).floor()]
                    + " "
                    + col_names[(cell_values[0] % num_cols)]).stable_diffusion()
            ),
            layout=weave.panels.BoardPanelLayout(x=0, y=0, w=24, h=12)
        ),
        weave.panels.BoardPanel(
            lambda col_names: weave.panels.Table(col_names, columns=[lambda col: col]),
            layout=weave.panels.BoardPanelLayout(x=0, y=12, w=12, h=6)
        ),
        weave.panels.BoardPanel(
            lambda row_names: weave.panels.Table(row_names, columns=[lambda row: row]),
            layout=weave.panels.BoardPanelLayout(x=12, y=12, w=12, h=6)
        ),
    ]
)
```
